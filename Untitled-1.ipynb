```python
# 获取数据

## 超参数
COLS_ALL = ['诉求分类', '业务类型', '受理时间', '一级类型', '二级类型', '受理内容']

import pandas as pd
import json

## 读取excel文件并获取指定列
df = pd.read_csv('Data\95598业务支持系统——诉求分类标记.csv', usecols=COLS_ALL, encoding='GBK')
print(df.tail())
```

```python
# 分词
import jieba
import jieba.posseg as pseg

## 增加
jieba.add_word('机器人服务')
jieba.add_word('主动抢修')
jieba.add_word('户号信息')
jieba.add_word('转人工')
jieba.add_word('更改联系方式')
## 对'受理内容'列进行分词
df['受理内容分词'] = df['受理内容'].apply(lambda x: " ".join(jieba.cut(x,HMM=True)))

print(df.head())
```

```python
#关键词抽取
import jieba.analyse

## 将所有'受理内容分词'的值合并为一个字符串
all_text = " ".join(df['受理内容分词'])

## 提取关键词
keywords = jieba.analyse.extract_tags(all_text, topK=100,  withWeight=True)

## print
for word in keywords:
    print("tag: %s\t\t weight: %f " % (word[0],word[1]))
```

```python
# 生成词云
import wordcloud
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

## 读取遮罩图片（可选）
mask = np.array(Image.open('Data/mask_img.jpg'))  # 替换为您的遮罩图片路径

## 创建词云对象
wc = wordcloud.WordCloud(
    font_path='simhei.ttf',  # 字体路径，根据实际情况修改
    mask=mask,  # 使用遮罩（可选）
    background_color='white'
)

## 生成词云
wc.generate(all_text)

## 显示词云
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
